```python
import pandas as pd 
wine_data=pd.read_csv("/home/dul/winequality-red.csv")
wine_data.head()
```

```python
# we get output as bool, saying True and False is of type bool
type(False)
type(True)
```

```python
# wine quality greater than 7 , append true to the list else append false
good_wine = []
for length in wine_data.quality:
    if length >= 7:
        good_wine.append(True)
    else:
        good_wine.append(False)
```

```python
# to display the list elements 0 to 9, here 10 is exclusive
good_wine[0:10]
```

```python
# no of rows, no. of columns in dataframe wine_data
wine_data.shape
```

```python
# no.of rows in good_wine list
len(good_wine)
# no. of rows in the dataframe and no. of rows in list is equal
```

```python
# to convert the python list to a pandas series
is_good = pd.Series(good_wine)
# view the good_wine series
is_good.head()
```

```python
# we want to get the wine where quality is greater than 7
wine_data[is_good]
# only wine quality greater than 7 is displayed
```

```python
# To completely remove the for loop
is_good = wine_data.quality >= 7
is_good.head()
# displayed the first 5 rows is_good value, if quality greater than 7, true
# if duration less than 7, displayed as boolean false
# pandas knows we want all the series of wine quality column attribute for each row compared
```

```python
# we want to get the wine where quality is greater than 7
wine_data[is_good]
# only wine with quality greater than 7 is displayed
```

```python
# generally the short form how every one writes
wine_data[wine_data.quality >= 7]
```
