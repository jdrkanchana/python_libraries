```python
import pandas as pd 
wine_data=pd.read_csv("/home/dul/winequality-red.csv")
wine_data.head()
```

```python
#Filter out both high wine quality and alcohol content in wine be 9.8
wine_data[(wine_data.quality>= 7) & (wine_data.alcohol == 9.8)]
```

```python
#Filter out both high wine quality or alcohol content in wine be 9.8
wine_data[(wine_data.quality>= 7) | (wine_data.alcohol == 9.8)]
```

```python
# the True or False output is displayed
# not the output filtered based of the table since square brackets were left off
(wine_data.quality>= 7) | (wine_data.alcohol == 9.8)
```

```python
# Filtering out based on switch case like scenario for one column attribute
wine_data[(wine_data.alcohol>= 9.80) |(wine_data.alcohol>= 10.00) |(wine_data.alcohol>= 11.20)]
```

```python
# Filtering out based on switch case like scenario for one column attribute
wine_data[wine_data.alcohol.isin([9.80,10.00,11.20])]
```
