```python
import pandas as pd
football=pd.read_csv("/home/dul/Football_dataset.csv")
football.tail()
```

```python
# if a row value is missing, the isnull() shows true
# if the row has a value and not a missing value, the isnull() shows false
football.isnull().tail()
```

```python
#notnull() is exact opposite of isnull()
# if a row value is missing, the notnull() shows false
# if the row has a value and not a missing value, the notnull() shows true
football.notnull().tail()
```

```python
# the count of the number of missing values in each row
# that is number of total rows having a missing value per each column
football.isnull().sum()
```
